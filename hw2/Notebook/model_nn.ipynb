```python
def load_data():
    #讀檔如果像這樣把路徑寫死交到github上去會馬上死去喔
    #還不知道怎寫請參考上面的連結
    x_train = pd.read_csv('./ml2019fall-hw2/data/X_train', )
    x_test = pd.read_csv('./ml2019fall-hw2/data/X_test')

    x_train = x_train.values
    x_test = x_test.values

    y_train = pd.read_csv('./ml2019fall-hw2/data/Y_train', header = None)
    y_train = y_train.values
    y_train = y_train.reshape(-1)

    return x_train, y_train, x_test

def sigmoid(z):
    res = 1 / (1.0 + np.exp(-z))
    return np.clip(res, 1e-6, 1-1e-6)
```

```python
def normalize(x_train, x_test):
    
    x_all = np.concatenate((x_train, x_test), axis = 0)
    mean = np.mean(x_all, axis = 0)
    std = np.std(x_all, axis = 0)

    index = [0, 1, 3, 4, 5,107]
    mean_vec = np.zeros(x_all.shape[1])
    std_vec = np.ones(x_all.shape[1])
    mean_vec[index] = mean[index]
    std_vec[index] = std[index]

    x_all_nor = (x_all - mean_vec) / std_vec

    x_train_nor = x_all_nor[0:x_train.shape[0]]
    x_test_nor = x_all_nor[x_train.shape[0]:]

    return x_train_nor, x_test_nor
```

```python
x_train, y_train, x_test = load_data()
x_train = np.column_stack((x_train, x_train[::,5]**2, 1/x_train[::,5]))
x_test = np.column_stack((x_test, x_test[::,5]**2, 1/x_test[::,5]))

x_train, x_test = normalize(x_train, x_test)
```

```python
from keras.utils.np_utils import to_categorical
y_train = to_categorical(y_train)
```

```python
# from sklearn.model_selection import train_test_split
# cv_X_train, cv_X_test, cv_y_train, cv_y_test = train_test_split(
#     x_train, y_train, test_size=0.5, random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
# from keras.layers import ActivityRegularization
from keras.layers import Dropout
from keras import regularizers

model = Sequential()
model.add(Dense(input_dim=108, output_dim=15,
          activation='sigmoid'))
# model.add(Dropout(rate=0.1, noise_shape=None, seed=None))

model.add(Dense(output_dim=10))
model.add(Activation('relu'))
# model.add(Dropout(rate=0.1, noise_shape=None, seed=None))

model.add(Dense(output_dim=5))
model.add(Dense(2, activation='softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=100, batch_size=300,)
```

```python
pred_y = model.predict(x_test)
pred_y = (pred_y[::,0] < 0.5).astype(int)

output = zip(list(range(1,len(pred_y)+1)), pred_y.ravel())
output = pd.DataFrame(output, columns = ['id','label'])
output_name = './nn.csv'
output.to_csv(output_name, index=False)
```

```python
def train(x_train, y_train):
    b = 0.0
    w = np.zeros(x_train.shape[1])
    lr = 0.05
    epoch = 1000
    b_lr = 0
    w_lr = np.ones(x_train.shape[1])
    
    for e in range(epoch):
        z = np.dot(x_train, w) + b
        pred = sigmoid(z)
        loss = y_train - pred

        b_grad = -1*np.sum(loss)
        w_grad = -1*np.dot(loss, x_train)

        b_lr += b_grad**2
        w_lr += w_grad**2


        b = b-lr/np.sqrt(b_lr)*b_grad
        w = w-lr/np.sqrt(w_lr)*w_grad

        if(e+1)%500 == 0:
            loss = -1*np.mean(y_train*np.log(pred+1e-100) + (1-y_train)*np.log(1-pred+1e-100))
            print('epoch:{}\nloss:{}\n'.format(e+1,loss))
    return w, b
```
